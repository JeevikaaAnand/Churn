```python
df = pd.read_csv("E:/Userfiles/Download/Churn.csv")
df
```

```python
df.info()
```

```python
df1 = df.copy()
df1["day.charge"] = df1["day.charge"].astype("float64")
df1["eve.mins"]= df1["eve.mins"].astype("float64")
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace = True)
```

```python
df1.drop(df1.columns[[0, 1, 2, 3]], axis=1, inplace=True)
df1
```

```python
df1.rename(columns= {'voice.plan' : 'voice_plan', 'voice.messages' : 'voice_messages', 'intl.plan':'intl_plan', 'intl.mins':'intl_mins', 'intl.calls':'intl_calls','intl.charge':'intl_charge', 'day.mins':'day_mins', 'day.calls': 'day_calls', 'day.charge':'day_charge', 'eve.mins':'eve_mins', 'eve.calls':'eve_calls', 'eve.charge':'eve_charge','night.mins':'night_mins', 'night.calls':'night_calls', 'night.charge': 'night_charge', 'customer.calls':'customer_calls'}, inplace = True)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df1['churn']=LE.fit_transform(df1['churn'])
```

```python
df1=pd.get_dummies(df1,columns=['voice_plan','intl_plan'], drop_first=True)
df1
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df1)
```

```python
X = df_norm.drop(['churn'], axis=1)
y = df_norm['churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
Xtrain.shape , Xtest.shape , ytrain.shape , ytest.shape 
```

```python
Xtrain
```

```python
from imblearn.over_sampling import RandomOverSampler
OS=RandomOverSampler(sampling_strategy=0.90)
Xtrain_os,ytrain_os=OS.fit_resample(Xtrain,ytrain)
```

```python
from collections import Counter
print("The number of classes before fit {}".format(Counter(ytrain)))
print("The number of classes after fit {}".format(Counter(ytrain_os)))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
Classifier = RandomForestClassifier()
Classifier.fit(Xtrain_os, ytrain_os)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
y_pred = Classifier.predict(Xtest)
print(confusion_matrix(ytest,y_pred))
print(accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
```

```python
prediction_proba = Classifier.predict_proba(Xtest)
prediction_proba
```

```python
input_data = (95, 20 ,5, 4, 300, 40, 55, 250, 50, 28, 300, 71, 50, 5, 1, 0)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = Classifier.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print('The person will not be churned')
else:
    print('The person will be churned')
```
